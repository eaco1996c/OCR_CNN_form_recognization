```python
import os, os.path
import numpy as np
from keras.preprocessing import image
import cv2
import re

from PIL import Image
import pytesseract as pt 
import pandas as pd 

# Import Keras for CNN model,
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Initialising the CNN
cnnmodel = Sequential()

cnnmodel.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
cnnmodel.add(MaxPooling2D(pool_size = (2, 2)))

cnnmodel.add(Conv2D(32, (3, 3), activation = 'relu'))
cnnmodel.add(MaxPooling2D(pool_size = (2, 2)))
   
cnnmodel.add(Flatten())    
 
cnnmodel.add(Dense(units = 128, activation = 'relu'))
cnnmodel.add(Dense(units = 3, activation = 'softmax'))

cnnmodel.compile(optimizer = 'rmsprop', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
train_gen = ImageDataGenerator(rescale = 1./255,
                               shear_range = 0.1,
                               zoom_range = 0.1,
                               horizontal_flip = True)

test_gen = ImageDataGenerator(rescale = 1./255)

train_set = train_gen.flow_from_directory('train_image',
                                           target_size = (64, 64),
                                           batch_size = 4,
                                           class_mode = 'binary')

test_set = test_gen.flow_from_directory('test_image',
                                         target_size = (64, 64),
                                         batch_size = 4,
                                         class_mode = 'binary')
```

```python
cnnmodel.fit(train_set,
             steps_per_epoch = 21,
             epochs = 50,
             validation_data = test_set,
             validation_steps = 3)

accuracy = cnnmodel.evaluate_generator(test_set, 100)

print('Accuracy of the model: ', accuracy)
```

```python
# Cut needed blocks for different information by ratios\n",
def imacrop(type,attribute,image):
    height,width = image.shape[:2]
    
    if(type=='f1'):
        if(attribute=='decedent'):
            startrow,startcol = int(height*0.1923),int(width *0.01)
            endrow,endcol = int(height *0.67),int(width *0.6591)            
            return image[startrow:endrow,startcol:endcol]
        
        elif(attribute=='date'):
            startrow,startcol = int(height*0.67307),int(width * 0.01)
            endrow,endcol = int(height * 0.99),int(width *0.2958)            
            return image[startrow:endrow,startcol:endcol]           
        
        elif(attribute=='section'):
            startrow,startcol = int(height*0.67307),int(width * 0.344)
            endrow,endcol = int(height *0.99 ),int(width *0.44)            
            return image[startrow:endrow,startcol:endcol]            
        
        elif(attribute=='lot'):
            startrow,startcol = int(height*0.67307),int(width * 0.4534)
            endrow,endcol = int(height *0.99),int(width *0.569)            
            return image[startrow:endrow,startcol:endcol]
        
        elif(attribute=='grave'):
            startrow,startcol = int(height*0.67307),int(width *0.627 )
            endrow,endcol = int(height * 0.99),int(width *0.6913)            
            return image[startrow:endrow,startcol:endcol]
        
    elif(type=='f2'):
        if(attribute=='decedent'):
            startrow,startcol = int(height*0.01),int(width *0.1136)
            endrow,endcol = int(height *0.306),int(width *0.99)            
            return image[startrow:endrow,startcol:endcol]
    
        elif(attribute=='date'):
            startrow,startcol = int(height*0.7551),int(width * 0.19)
            endrow,endcol = int(height * 0.99),int(width *0.99)            
            return image[startrow:endrow,startcol:endcol]           
        
        elif(attribute=='section'):
            startrow,startcol = int(height*0.306),int(width * 0.647)
            endrow,endcol = int(height *0.51 ),int(width *0.99)            
            return image[startrow:endrow,startcol:endcol]           
        
        elif(attribute=='lot'):
            startrow,startcol = int(height*0.306),int(width * 0.1136)
            endrow,endcol = int(height *0.51 ),int(width *0.4525)            
            return image[startrow:endrow,startcol:endcol]
        
        elif(attribute=='grave'):
            startrow,startcol = int(height*0.51),int(width *0.1136)
            endrow,endcol = int(height * 0.7551),int(width *0.4525)            
            return image[startrow:endrow,startcol:endcol]
        
    elif(type=='f3'):
        if(attribute=='decedent'):
            startrow,startcol = int(height*0.001),int(width *0.0935)
            endrow,endcol = int(height *0.3857),int(width *0.9)            
            return image[startrow:endrow,startcol:endcol]
        
        elif(attribute=='date'):
            startrow,startcol = int(height*0.85714),int(width * 0.1613)
            endrow,endcol = int(height * 0.99),int(width *0.9)            
            return image[startrow:endrow,startcol:endcol]           
       
        elif(attribute=='section'):
            startrow,startcol = int(height*0.3857),int(width * 0.5806)
            endrow,endcol = int(height *0.6 ),int(width *0.9)           
            return image[startrow:endrow,startcol:endcol]            
     
        elif(attribute=='lot'):
            startrow,startcol = int(height*0.3857),int(width * 0.0935)
            endrow,endcol = int(height *0.6 ),int(width *0.4516)            
            return image[startrow:endrow,startcol:endcol]
      
        elif(attribute=='grave'):
            startrow,startcol = int(height*0.6),int(width *0.0935 )
            endrow,endcol = int(height * 0.85),int(width *0.4516)            
            return image[startrow:endrow,startcol:endcol]
```
